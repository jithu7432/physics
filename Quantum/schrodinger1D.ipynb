```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.linalg import eigh_tridiagonal as et
from matplotlib.animation import FuncAnimation
from matplotlib import style
import sympy as sp

style.use('default')
```

```python
def potential(x):
    return 1000 * (x - 0.5)**2
```

```python
N = 2000
dy = 1 / N
y = np.linspace(0, 1, N + 1)
```

```python
V = potential(y)
d =  1 / dy**2 + potential(y)[1:-1]
e = -1 / (2 * dy**2) * np.ones(len(d) - 1)
w, v = et(d, e)
```

```python
fig, axes = plt.subplots(1,1,figsize = (8,4))
axes.set_xlim(0,1)
axes.set_xlabel(r'L')
axes.set_ylabel(r'$\psi$')
for i in range(5):
    axes.plot(y[1:-1],v.T[i],label = i)
axes.grid()
axes.legend()
```
